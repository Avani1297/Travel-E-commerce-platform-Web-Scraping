```python
import os
import requests
import re

driver = webdriver.Chrome()

driver.implicitly_wait(120)
driver.set_script_timeout(120)
driver.set_page_load_timeout(120)

# Create a folder for HTML files
folder = "DDR Final Project"
os.makedirs(folder, exist_ok=True)

# Get HTML for each url and save it in the folder
for index, complete_url in enumerate(complete_urls, 1):
    driver.get(complete_url)
    time.sleep(5)

    filename = os.path.join(folder, f"{index}.html")
    
    with open(filename, 'w', encoding='utf-8') as file:
        file.write(driver.page_source)
```

```python
folder = "DDR Final Project"

GetYourGuide = []

# Loop through each file in the directory
for filename in os.listdir(folder):

    # Check if the file ends with .html
    if filename.endswith(".html"):

        print(filename)

        # Construct the full file path
        filepath = os.path.join(folder, filename)

        # Read file to string
        with open(filepath, 'r', encoding='utf-8') as file:
             html = file.read()

        # Use html string
        soup = BeautifulSoup(html, 'html.parser')

        time.sleep(1)

        country = soup.select_one('li.activity-breadcrumbs__item').text.strip()
        print('Country:', country)

        location1 = soup.select('li.activity-breadcrumbs__item')[1].text.strip()
        print('Location 1:', location1)

        location2 = soup.select('li.activity-breadcrumbs__item')[2].text.strip()
        if location2:
            location2 = re.sub(r'Things to do in ','', location2)
        print('Location 2:', location2)

        location3 = soup.select('li.activity-breadcrumbs__item')[3].text.strip()
        location3 = "NULL" if location3 == country else location3
        print('Location 3:', location3)

        category = soup.select_one('span.c-classifier-badge').text.strip()
        print('Product Category:', category)

        name = soup.select_one('h1.activity__title').text.strip()
        print('Product Name:', name)

        review_score_element = soup.select_one('span.reviews-summary__rating-average--rating')
        review_score = review_score_element.text.strip() if review_score_element else "NULL"
        print('Review Score:', review_score)

        review_count_element = soup.select_one('span.activity-rating--reviews-count')
        review_count = re.sub(r'[rR]eviews?', '', review_count_element.text.strip()) if review_count_element else "NULL"
        print('Review Count:', review_count)

        activity_provider_element = soup.select_one('div.supplier-name')
        activity_provider = re.sub(r'.*:\s*', '', activity_provider_element.text.strip()) if activity_provider_element else "NULL"
        print('Activity Provider:', activity_provider)

        cancellation_policy = soup.select_one('dt.activity-key-details__term').text.strip()
        print('Cancellation Policy:', cancellation_policy)

        cancellation_policy_details = soup.select_one('dd.activity-key-details__description').text.strip()
        print('Cancellation Policy Details:', cancellation_policy_details)

        promotion_tag_element = soup.select_one('span.c-marketplace-badge')
        promotion_tag = promotion_tag_element.text.strip() if promotion_tag_element else "NULL"
        print('Promotion Tag:', promotion_tag)

        price_element = soup.select_one('span.baseline-pricing__from--value')
        if not price_element:
            price_element = soup.select_one('strong.price-block__price-actual')
        price = price_element.text.strip() if price_element else "NULL"
        if price:
             price = re.sub(r'\$', '', price)
        print('Price:', price, '\n')


        GetYourGuide.append({
            'Country': country,
            'Location 1': location1,
            'Location 2': location2,
            'Location 3': location3,
            'Product Category': category,
            'Product Name': name,
            'Review Score': review_score,
            'Review Count': review_count,
            'Activity Provider': activity_provider,
            'Cancellation Policy': cancellation_policy,
            'Cancellation Policy Details': cancellation_policy_details,
            'Promotion Tag': promotion_tag,
            'Price': price
            
        })

        time.sleep(10)

# Create a DataFrame from the list of dictionaries
final_df = pd.DataFrame(GetYourGuide)

# Display the final DataFrame
print(final_df)
```

```python
final_df.to_csv('GetYourGuide.csv', index=False)
```
